```python
import functools
import operator
import itertools
import math
import sys
import json
import re
from enum import Enum
from collections import deque
from dataclasses import dataclass

import pulp

data = open("sample_input.txt").read().splitlines()
original = [int(line) for line in data]

class Node():
    def __init__(self, index, value):
        self.value = value
        self.original_index = index
        self.index = index
        self.prev = None
        self.next = None
    def __repr__(self) -> str:
        return f"{self.prev.value} <- {self.value}@{self.index} -> {self.next.value}"
    def mix(self):
        if self.value == 0:
            return
        if self.value > 0:
            left = self.next
            right = left.next
            for _ in range(0, self.value - 1):
                left = right
                right = left.next
        else:
            right = self.prev
            left = right.prev
            for _ in range(0, self.value + 1, -1):
                right = left
                left = right.prev
        
        re_index_from = self.next if self.index == 0 else None

        # remove self from list
        self.prev.next, self.next.prev = self.next, self.prev

        # insert self between left and right
        left.next = self
        right.prev = self
        self.prev = left
        self.next = right

        if re_index_from:
            current = re_index_from.next
            re_index_from.index = 0
            index = 1
            while current.original_index != re_index_from.original_index:
                current.index = 1
                current = current.next

nodes = [Node(i,v) for i,v in enumerate(original)]

node_zero = None

for i,n in enumerate(nodes):
    next_i = i + 1
    if next_i == len(nodes):
        next_i = 0
    n.next = nodes[next_i]
    nodes[next_i].prev = n
    if n.value == 0:
        node_zero = n

def find_index_zero():
    for n in nodes:
        if n.index == 0:
            return n

def render():
    zero = find_index_zero()
    ls = [zero.value]
    current = zero.next
    while current != zero:
        ls.append(current.value)
        current = current.next
    return ls

for i,n in enumerate(nodes):
    n.mix()

def find_wrapped_ordinal(n):
    current = node_zero
    for _ in range(0, n):
        current = current.next
    return current.value

a = find_wrapped_ordinal(1000)
b = find_wrapped_ordinal(2000)
c = find_wrapped_ordinal(3000)
print(a)
print(b)
print(c)

print(a+b+c)


```
