```python
import functools
import operator
import itertools
import math
import sys
import json
import re
from enum import Enum
from collections import deque
from dataclasses import dataclass
from typing import Dict,List,Tuple,TypeAlias

src = "sample_input_2.txt"
data = open("sample_input_2.txt").read().splitlines()

def add(a,b):
    return tuple([x + y for x,y in zip(a,b)])

def mul(a, n): 
    return tuple([x * n for x in a])

C3: TypeAlias = Tuple[int,int,int]

rotations = {
    "U":()
}

class Cube():
    def __init__(self, s: int, texture: List[List[str]]) -> None:
        self.faces = [[["."] * s] *s]*s
        self.corners = {
            "a":(0,0,0),
            "b":(s,0,0),
            "c":(s,s,0),
            "d":(0,s,0),
            "e":(0,0,s),
            "f":(s,0,s),
            "g":(s,s,s),
            "h":(0,s,s),
        }
    def






class Cell():
    def __init__(self, coords: Tuple[int,int], state: str) -> None:
        self.coords = coords
        self.row, self.col = coords
        self.passable = state == "."
        self.state = state
        self.neighbors: Dict[Tuple[int,int], "Cell"] = {}
    def __repr__(self) -> str:
        return f"{self.coords}:{self.state}"

cells: List[Cell] = []
cell_map: Dict[Tuple[int,int],Cell] = {}
height = len(data)
width = 0
for r, row in enumerate(data):
    if row == "":
        break
    for c, col in enumerate(row):
        if col == " ":
            continue
        cell = Cell((r,c), col)
        cells.append(cell)
        width = max(width,c)
        cell_map[(r,c)] = cell

max_r = height - 1
max_c = width - 1

facings = {
    (-1,0):"^",
    (1,0):"v",
    (0,1):">",
    (0,-1):"<",
}

directions = [
    (0,1),
    (1,0),
    (0,-1),
    (-1,0),
]

def add(a,b):
    return (a[0]+b[0],a[1]+b[1])

def within(coords: Tuple[int,int]) -> bool:
    r,c = coords
    return r >= 0 and r <= max_r and c >= 0 and c <= max_c

def link(cell: Cell):
    for d in directions:
        loc = add(cell.coords,d)
        while not (neighbor := cell_map.get(loc)):
            r,c = add(loc,d)
            if r < 0:
                r = max_r
            elif r > max_r:
                r = 0
            elif c < 0:
                c = max_c
            elif c > max_c:
                c = 0
            loc = (r,c)
        
        cell.neighbors[d] = neighbor

for cell in cells:
    link(cell)
        
print(f"start: {cells[0]}")
print(f"start neighbors: {cells[0].neighbors.values()}")

pos = cells[0].coords
facing = 0
def get_dir(f:int) -> Tuple[int,int]:
    return directions[f]
def get_dir_repr(f:int) -> str:
    return facings[get_dir(f)]
def rotate(f: int, turn: str) -> int:
    if turn == "R":
        return (f + 1) % 4
    else:
        return (f + 3) % 4
history = {
    pos: get_dir_repr(facing)
}


def dump():
    for r in range(0, height):
        for c in range(0,width):
            f = history.get((r,c))
            if f:
                print(f, end = "")
                continue
            cell = cell_map.get((r,c), None)
            if cell:
                print(cell.state, end = "")
            else:
                print(" ", end = "")
        print()

dump()

moves = [int(x) if x.isnumeric() else x for x in re.findall("(\d+|\D)", data[-1])]

for m in moves:
    #print(f"{m}: ", end="")
    if isinstance(m,int):
        cell = cell_map[pos]
        dir = get_dir(facing)
        for _ in range(0, m):
            next = cell.neighbors.get(dir)
            if not next:
                #print(f"!!! cell {cell} has no neighbor at {dir}")
                continue
            if next.passable:
                cell = next
                pos = cell.coords
                #print(f" moved to {pos}", end = ";")
                history[pos] = get_dir_repr(facing)
            else:
                break
    else:
        #print(f"rotating {m}", end = ";")
        facing = rotate(facing, m)
        history[pos] = get_dir_repr(facing)
    #print()
    #dump()


final_pos = add((1,1), pos)
print(f"part 1: facing {facing}=>{get_dir_repr(facing)} at {final_pos}")

print(1000 * final_pos[0] + 4 * final_pos[1] + facing)
```
