```python
import functools
import operator
import itertools
import math
import sys
import json
import re

data = open("input.txt").read()
lines = []
min_x = math.inf
min_y = 0
max_x = 0
max_y = 0

occupied = {}

rock_chains = []
for line in data.splitlines():
    nums = [int(n) for n in re.sub("\D", "\n", line).split()]
    coords = [(nums[i], nums[i+1]) for i in range(0, len(nums), 2)]
    chain = list(itertools.pairwise(coords))
    for start, end in chain:
        sx, sy = start
        ex, ey = end
        if ex < sx or ey < sy:
            sx, sy = end
            ex, ey = start
        for x in range(sx, ex + 1):
            for y in range(sy, ey + 1):
                occupied[(x,y)] = "#"
                max_y = max(max_y, y)
                min_x = min(min_x, x)
                max_x = max(max_x, x)

print(rock_chains)

def show():
    for y in range(min_y, max_y + 1):
        print(f"{y:2} ", end = "")
        for x in range(min_x, max_x + 1):
            print(occupied.get((x,y), "."), end = "")
        print()

moves = [
    (0,1),
    (-1,1),
    (1,1)
]

max_y = max_y + 2
min_x = min_x - 1000
max_x = max_x + 1000
for x in range(min_x, max_x):
    occupied[(x,max_y)] = "#"

show()

done = False
count = 0
while True:
    count += 1
    x = 500
    y = 0
    settled = False
    while y <= max_y and not settled:
        nx = x
        ny = y
        for dx, dy in moves:
            nx = x + dx
            ny = y + dy
            v = occupied.get((nx,ny), ".")
            if v == ".":
                x = nx
                y = ny   
                break
        else:
            occupied[(x,y)] = "o"
            settled = True
    #print(count)
    #show()
    if y == 0:
        break

print(count)


```
